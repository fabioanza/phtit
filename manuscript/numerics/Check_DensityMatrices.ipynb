```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
phi = 3*np.pi/4
HH = 1/np.sqrt(2)*np.array([[1,1],[1,-1]])
UU = np.array([[1,0],[0,np.exp(1j*phi)]])
```

```python
def Circuit(phi):
    HH = 1/np.sqrt(2)*np.array([[1,1],[1,-1]])
    UU = np.array([[1,0],[0,np.exp(1j*phi)]])
    return np.matmul(HH,np.matmul(UU,HH))
```

```python
def Map(MM,rho):
    return np.matmul(MM,np.matmul(rho,MM.conj().T))
```

```python
def Alice_PostCircuit(lam,phi):
    p_AH = (1+lam)/2
    p_AT = 1 - p_AH
    rho_AH = np.array([[1,0],[0,0]])
    rho_AT = np.array([[0,0],[0,1]])
    MM = Circuit(phi)
    return p_AH*Map(MM,rho_AH)+p_AT*Map(MM,rho_AT)
```

```python
def Bob_PostCircuit(lam,chi,phi):
    p_BH = 1/2
    p_BT = 1 - p_BH
    rho_BH = np.array([[(1+lam)/2,np.sqrt((1-lam**2)/4)*np.exp(-1j*chi)],[np.sqrt((1-lam**2)/4)*np.exp(1j*chi),(1-lam)/2]])
    rho_BT = np.array([[(1+lam)/2,-np.sqrt((1-lam**2)/4)*np.exp(-1j*chi)],[-np.sqrt((1-lam**2)/4)*np.exp(1j*chi),(1-lam)/2]])
    MM = Circuit(phi)
    return p_BH*Map(MM,rho_BH)+p_BT*Map(MM,rho_BT)
```

```python
lam = 0.5
chi = np.pi/2
PHI = np.linspace(0,2*np.pi,100)
P0_Alice = np.zeros(len(PHI),dtype=complex)
P0_Bob = np.zeros(len(PHI),dtype=complex)
for k in range(len(PHI)):
    P0_Alice[k] = Alice_PostCircuit(lam,PHI[k])[0,0]
    P0_Bob[k] = Bob_PostCircuit(lam,chi,PHI[k])[0,0]
```

```python
plt.plot(PHI,np.real(P0_Alice),PHI,np.real(P0_Bob))
plt.show()
```

```python
plt.plot(PHI,np.imag(P0_Alice),PHI,np.imag(P0_Bob))
plt.plot()
```
