```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Alice_1(delta_phi,lam,chi):
    return 0.5+0.5*lam*np.cos(delta_phi)

def Alice_2(delta_phi,lam,chi):
    return 0.5-0.5*lam*np.cos(delta_phi)

def Bob_1(delta_phi,lam,chi):
    return 0.5+0.5*lam*np.cos(delta_phi)-np.sqrt((1-lam**2)/4)*np.sin(delta_phi)*np.sin(chi)

def Bob_2(delta_phi,lam,chi):
    return 0.5-0.5*lam*np.cos(delta_phi)+np.sqrt((1-lam**2)/4)*np.sin(delta_phi)*np.sin(chi)
```

```python
DPHI = np.linspace(0,2*np.pi,100)
lam = 0.5
chi = np.pi/2
A1 = np.zeros(len(DPHI))
A2 = np.zeros(len(DPHI))
B1 = np.zeros(len(DPHI))
B2 = np.zeros(len(DPHI))
for k in range(len(DPHI)):
    A1[k] = Alice_1(DPHI[k],lam,chi)
    A2[k] = Alice_2(DPHI[k],lam,chi)
    B1[k] = Bob_1(DPHI[k],lam,chi)
    B2[k] = Bob_2(DPHI[k],lam,chi)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(DPHI,A1,label='Alice',linewidth=5)
ax.plot(DPHI,B1,label='Bob',linewidth=5)
ax.legend(fontsize=30)
ax.set_ylim([0,1.01])
ax.set_title('Detector 1 - Probability'.format(lam,chi),fontsize=40)
ax.set_xlabel('$\phi_0-\phi_1$',fontsize=30)
ax.tick_params(axis='both', which='major', labelsize=20)
plt.xticks([0, np.pi/2, np.pi,3*np.pi/2, 2*np.pi], ['0', '$\pi/2$', '$\pi$','$3\pi/2$','$2\pi$'],rotation=20)

plt.savefig('./Plot/Distinguishability.png')

plt.show()
```
