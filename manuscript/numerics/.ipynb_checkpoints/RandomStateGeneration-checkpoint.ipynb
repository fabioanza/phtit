```python
import numpy as np
import matplotlib.pyplot as plt
import random
from mpl_toolkits.mplot3d import Axes3D
```

```python
def boundaries_check(point):
    if point[0]>=0 and point[0]<=1 and point[1]>=-np.pi and point[1]<=np.pi:
        return True
    else:
        return False
```

```python
def gaussian2d(mean,cov,sample_size):
    x = 100*np.ones(sample_size)
    y = 100*np.ones(sample_size)
    s = 0
    while s < sample_size:
        aux = np.random.multivariate_normal(mean, cov)
        if boundaries_check(aux)==True:
            x[s], y[s] = aux
            s = s+1

    return x,y 
```

```python
def exponential(beta,sample_size):
    x = 100*np.ones(sample_size)
    y = 100*np.ones(sample_size)
    s = 0
    while s < sample_size:
        aux1 = np.random.exponential(beta)
        aux2 = (2*np.random.rand()-1)*np.pi
        if boundaries_check(np.array([aux1,aux2]))==True:
            x[s] = aux1
            y[s] = aux2
            s = s+1

    return x,y 
```

```python
def dirac_delta(x0,y0,sample_size):
    x = x0*np.ones(sample_size)
    y = y0*np.ones(sample_size)
    return x,y 
```

```python
def colorplot_distribution(sample,Mx,My):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    x, y = sample
    hist, xedges, yedges = np.histogram2d(x, y, bins=(Mx,My) , range=[[0, 1], [-np.pi, np.pi]])
    pesi = 1/len(x)*np.ones(len(x))
    plt.hist2d(x,y, bins=(20,20), range=[[0, 1], [-np.pi, np.pi]], weights=pesi, cmap='YlOrBr')
    cb = plt.colorbar()
    cb.set_label('probability')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    
    #X-axis goes from 0 to 1
    # Major xticks every 5/M, minor xticks every 1/M
    major_xticks = np.arange(0, 1.01, 5/Mx)
    minor_xticks = np.arange(0, 1.01, 1/Mx)

    #Y-axis goes from 0 to 2pi
    # Major yticks every 2pi*5/M, minor xticks every 2pi*/M
    major_yticks = np.arange(-np.pi, np.pi+0.01, 2*np.pi*5/My)
    minor_yticks = np.arange(-np.pi, np.pi+0.01, 2*np.pi/My)

    ax.set_xticks(major_xticks)
    ax.set_xticks(minor_xticks, minor=True)
    #x_label = [r"$0$", r"$\frac{1}{4}$", r"$\frac{1}{2}$", r"$\frac{3}{2}$", r"$1$"]
    #ax.set_xticklabels(x_label, fontsize=20)

    ax.set_yticks(major_yticks)
    ax.set_yticks(minor_yticks, minor=True)
    #y_label = [ r"$-\pi$",r"$-\frac{1}{2}\pi$",r"$0$", r"$\frac{1}{2}\pi$", r"$\pi$"]
    #ax.set_yticklabels(y_label, fontsize=20)

    # Or if you want different settings for the grids:
    ax.grid(which='minor', alpha=0.2)
    ax.grid(which='major', alpha=1)
    
```

```python
def histogramplot_distributions(sample):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    x, y = sample
    hist, xedges, yedges = np.histogram2d(x, y, bins=20, range=[[0, 1], [-np.pi, np.pi]])
    
    # Construct arrays for the anchor positions of the 16 bars.
    xpos, ypos = np.meshgrid(xedges[:-1] + (xedges[1]-xedges[0])/2, yedges[:-1] + (yedges[1]-yedges[0])/2, indexing="ij")
    xpos = xpos.ravel()
    ypos = ypos.ravel()
    zpos = 0
    
    # Construct arrays with the dimensions for the 16 bars.
    dx = (xedges[1]-xedges[0])
    dy = (yedges[1]-yedges[0])
    dz = hist.ravel()
    ax.set_xlabel('x')
    ax.set_ylabel('y')


    #X-axis goes from 0 to 1
    # Major xticks every 0.2
    major_xticks = np.arange(0, 1.02, 0.25)
    minor_xticks = np.arange(0, 1, 0.05)

    #Y-axis goes from -pi to pi
    #Major yticks every 2pi/5
    major_yticks = np.arange(-np.pi, np.pi+0.01, 2*np.pi/4)
    minor_yticks = np.arange(-np.pi, np.pi+0.01, 2*np.pi/20)

    ax.set_xticks(major_xticks)
    ax.set_xticks(minor_xticks, minor=True)
    x_label = [r"$0$", r"$0.25$", r"$0.5$", r"$0.75$", r"$1$"]
    ax.set_xticklabels(x_label, fontsize=10)

    ax.set_yticks(major_yticks)
    ax.set_yticks(minor_yticks, minor=True)
    y_label = [ r"$-\pi$",r"$-\frac{\pi}{2}$",r"$0$", r"$\frac{\pi}{2}$", r"$\pi$"]
    ax.set_yticklabels(y_label, fontsize=10)

    # Or if you want different settings for the grids:
    ax.grid(which='minor', alpha=0.2)
    ax.grid(which='major', alpha=1)

    ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')
    
```

```python
mean = np.array([0.1,np.pi/2])
cov = np.array([[0.01,0],[0,0.01]])
sample_size = 10000
sample_gaussian = gaussian2d(mean,cov,sample_size)
beta = 5
sample_exponential = exponential(1/beta,sample_size)
sample_diracdelta = dirac_delta(0.63,np.pi/4,sample_size)
```

```python
np.save('./Data/RandomStates_gaussian',sample_gaussian)
np.save('./Data/RandomStates_exponential',sample_exponential)
np.save('./Data/RandomStates_diracdelta',sample_diracdelta)
```

```python
colorplot_distribution(sample_diracdelta,20,20)
colorplot_distribution(sample_exponential,20,20)
colorplot_distribution(sample_gaussian,20,20)
```

```python
histogramplot_distributions(sample_diracdelta)
histogramplot_distributions(sample_exponential)
histogramplot_distributions(sample_gaussian)
```
